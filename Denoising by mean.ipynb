```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
import skimage.io
import skimage.morphology
```

```python
def Denoising(path):
    
    originalimage=skimage.io.imread(path)
    image_noisy = skimage.util.random_noise(originalimage, mode='s&p')
    filtered_median = skimage.filters.median(image_noisy)

    plt.subplots(figsize=(20,20))
    plt.subplot(1,3,1)
    plt.title('Orginal')
    plt.imshow(originalimage)


    plt.subplot(1,3,2)
    plt.title('Noisy')
    plt.imshow(image_noisy)


    plt.subplot(1,3,3)
    plt.title('Denoised')
    plt.imshow(filtered_median) 
    
    plt.show()
```

```python
path='....\\Desktop\\Stuff\\cover.jpg'
Denoising(path)
```
